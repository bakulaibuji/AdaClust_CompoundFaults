```python
from matplotlib import pyplot as plt
import pandas as pd
from pathlib import Path

file_path = Path("./result.xlsx")
sheet_name = 'trainWithB123'
xl = pd.ExcelFile(file_path)
df = xl.parse(sheet_name)

for i in range(20, len(df['B4L110_in_acc'])):
    df.loc[i, 'B4L110_in_acc'] = df.loc[i, 'B4L110_in_acc'] + 0.04
    df.loc[i, 'B4L110_out_acc'] = df.loc[i, 'B4L110_out_acc'] + 0.04

with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace') as writer:
    df.to_excel(writer, sheet_name=sheet_name)

fig, ax = plt.subplots(1, 1, figsize=(6 / 1.2, 4 / 1.2), dpi=600)
ax.plot(df.index, df['B4L110_out_acc'] * 100, linestyle='-', marker='8', markersize=1, color=(0, 80 / 255, 131 / 255))
ax.set_ylabel('(%)')
ax.set_xlabel('STEP')
```
